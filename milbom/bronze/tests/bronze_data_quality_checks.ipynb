```python
from pyspark.sql.functions import current_timestamp, current_date, lit, col, max as spark_max
```

```python
df = spark.table("milbom_bronze.bakehouse_sales_transactions")

# Null checks
invalid_nulls = df.filter(
    col("transactionID").isNull() |
    col("customerID").isNull() |
    col("franchiseID").isNull()
)

# Duplicate transactions
duplicate_ids = (
    df.groupBy("transactionID")
    .count()
    .filter(col("count") > 1)
)

# Price consistency
invalid_prices = df.filter(col("totalPrice") != col("quantity") * col("unitPrice"))
```
