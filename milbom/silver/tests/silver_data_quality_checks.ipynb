```python
from pyspark.sql.functions import col, count
```

```python
df = spark.table("milbom_silver.bakehouse_sales_transactions")

# Primary Key validation
assert df.filter(col("transactionID").isNull()).count() == 0

# Financial consistency
invalid_amounts = df.filter(col("totalPrice") != col("quantity") * col("unitPrice"))
```

```python
df = spark.table("milbom_silver.bakehouse_sales_points")

# PK validation
assert df.filter(col("salesPointsID").isNull()).count() == 0

# Geolocation check
invalid_geo = df.filter(
    (col("latitude").isNull()) | (col("longitude").isNull())
)
```

```python
df = spark.table("milbom_silver.bakehouse_customers")

# PK validation
assert df.filter(col("customerID").isNull()).count() == 0

# Gender domain validation
invalid_gender = df.filter(~col("gender").isin("Male", "Female", "Other"))
```
